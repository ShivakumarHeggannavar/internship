Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and 
salary filter.  
You have to scrape data for “Data Scientist” designation for first 10 job results.  
You have to scrape the job-title, job-location, company name, experience required.  
The location filter to be used is “Delhi/NCR”. The salary filter to be used is “3-6” lakhs 

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.naukri.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys("Data Scientist")
```

```python
locations=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
locations.send_keys("Delhi/NCR")
```

```python
search=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[6]")
search.click()
```

```python
loc_filter=driver.find_elements(By.XPATH,'//span[@class="locWdth"]')

for i in loc_filter :
    if(i.text=="Delhi/NCR") :
        i.click()
        break
```

```python
salary=driver.find_elements(By.XPATH,'//div[@class="styles_filterOptns__1vq77"]')
for i in salary:
    if(i.text=='3-6 lakhs'):
        i.click()
        break
```

```python
job_title=[]
title=driver.find_elements(By.XPATH,'//div[@class=" row1"]')
for i in title:
    job_title.append(i.text)

job_title[10:20]
```

```python
company_names=[]
company_name=driver.find_elements(By.XPATH,'//a[@class=" comp-name mw-25"]')

for i in company_name:
    company_names.append(i.text)
company_names[:10]
```

```python
location=[]
loc=driver.find_elements(By.XPATH,'//span[@class="locWdth"]')

for i in loc:
    location.append(i.text)
location[10:20]
```

```python
experience=[]
exp=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in exp:
    experience.append(i.text)
experience[10:20]
```

```python
import pandas as pd
df1=pd.DataFrame({"JOB_TITLE":job_title[10:20],"LOCATION":location[10:20],"COMPANY_NAME":company_names[:10],"EXPERIENCE":experience[10:20]})
```

```python
df1
```

Q6:Write a python program to scrape data for Top 1000 Quotes of All Time. 
The above task will be done in following steps: 
1. First get the webpagehttps://www.azquotes.com/ 
2. Click on Top Quote 
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.azquotes.com/")
```

```python
top_quotes=driver.find_element(By.XPATH,'//a[@href="/top_quotes.html"]')
top_quotes.click()
```

```python
Authors=[]
```

```python
start=0
end=10
for i in range(start,end):
    author=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author:
        Authors.append(i.text)
    next_button=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a")
    next_button.click()
    time.sleep(5)
Authors[:1000]
```

```python
quotes=[]
```

```python
start=0
end=10
for i in range(start,end):
    quote=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote:
        quotes.append(i.text)
    next_button=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a")
    next_button.click()
    time.sleep(5)
quotes[:1000]
```

```python
print(len(Authors[:1000]))
```

```python
df6=pd.DataFrame({"Authors":Authors[:1000],"Quotes":quotes[:1000]})
```
